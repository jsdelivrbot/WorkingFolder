```python
chicago_crime['Crime_Group'] = ''
```

```python
chicago_crime_2017 = chicago_crime.loc[chicago_crime['Year'] == 2017]
```

```python
chicago_crime_2017_robb = chicago_crime_2017.loc[chicago_crime_2017['Community Area'] == 'ROBBERY']
chicago_crime_2017_robb
```

```python
def crimegr(a):
    
      if a['Primary Type'] == 'BURGLARY':
        return 'THEFT'
      elif a['Primary Type'] == 'HOMICIDE':
        return 'HOMICIDE'
      elif a['Primary Type'] == 'ROBBERY':
        return 'THEFT'
      elif a['Primary Type'] == 'THEFT':
        return 'THEFT'
      elif a['Primary Type'] == 'MOTOR VEHICLE THEFT':
        return 'THEFT'
      elif a['Primary Type'] == 'CRIMINAL TRESPASS':
        return 'THEFT'
      elif a['Primary Type'] == 'BATTERY':
        return 'BATTERY'
      elif a['Primary Type'] == 'OTHER OFFENSE':
        return 'BATTERY'
      elif a['Primary Type'] == 'ASSAULT':
        return 'BATTERY'
      elif a['Primary Type'] == 'KIDNAPPING':
        return 'BATTERY'
      elif a['Primary Type'] == 'ARSON':
        return 'BATTERY'
      elif a['Primary Type'] == 'INTIMIDATION':
        return 'BATTERY'
      elif a['Primary Type'] == 'DECEPTIVE PRACTICE':
        return 'DECEPTIVE PRACTICE'
      elif a['Primary Type'] == 'CONCEALED CARRY LICENCE VIOLATION':
        return 'WEAPON VIOLATION'
      elif a['Primary Type'] == 'CRIMINAL SEXUAL ASSAULT':
        return 'SEXUAL'
      elif a['Primary Type'] == 'SEX OFFENSE':
        return 'SEXUAL'
      elif a['Primary Type'] == 'HUMAN TRAFFICKING':
        return 'SEXUAL'
      elif a['Primary Type'] == 'PROSTITUTION':
        return 'SEXUAL'
      elif a['Primary Type'] == 'OFFENSE INVOLVING CHILDREN':
        return 'SEXUAL'
      elif a['Primary Type'] == 'NARCOTICS':
        return 'NARCOTICS'
      elif a['Primary Type'] == 'GAMBLING':
        return 'NARCOTICS'
      elif a['Primary Type'] == 'LIQUOR LAW VIOLATION':
        return 'NARCOTICS'
      elif a['Primary Type'] == 'OTHER NARCOTIC VIOLATION':
        return 'NARCOTICS'    
      elif a['Primary Type'] == 'PUBLIC PEACE VIOLATION':
        return 'NON-CRIMINAL'
      elif a['Primary Type'] == 'OBSCENITY':
        return 'NON-CRIMINAL'
      elif a['Primary Type'] == 'RITUALISM':
        return 'NON-CRIMINAL'
      elif a['Primary Type'] == 'PUBLIC INDECENCY':
        return 'NON-CRIMINAL'
      elif a['Primary Type'] == 'INTERFERENCE WITH PUBLIC OFFICER':
        return 'NON-CRIMINAL'
      elif a['Primary Type'] == 'NON-CRIMINAL':
        return 'NON-CRIMINAL'

chicago_crime_2017['Crime_Group'] = chicago_crime_2017.apply(crimegr, axis=1)
```

```python
table_crime_2017_crimegroups = pd.pivot_table(chicago_crime_2017, values='ID', index=['Community Area'],columns=['Crime_Group'], aggfunc=np.count_nonzero)
table_crime_2017_crimegroups.fillna(0)
```

```python
table_crime_2017_crimegroups.to_csv('CrimeType2017.csv')
```
