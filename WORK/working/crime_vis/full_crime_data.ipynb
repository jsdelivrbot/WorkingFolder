```python
import pandas as pd
import numpy as np
```

```python
crimes = pd.read_csv("Crimes_-_2001_to_present.csv").fillna(0)
crimes.head()
```

```python
def crimegr(a):
    
      if a['Primary Type'] == 'BURGLARY':
        return 'THEFT'
      elif a['Primary Type'] == 'HOMICIDE':
        return 'HOMICIDE'
      elif a['Primary Type'] == 'ROBBERY':
        return 'THEFT'
      elif a['Primary Type'] == 'THEFT':
        return 'THEFT'
      elif a['Primary Type'] == 'MOTOR VEHICLE THEFT':
        return 'THEFT'
      elif a['Primary Type'] == 'CRIMINAL TRESPASS':
        return 'THEFT'
      elif a['Primary Type'] == 'BATTERY':
        return 'BATTERY'
      elif a['Primary Type'] == 'OTHER OFFENSE':
        return 'BATTERY'
      elif a['Primary Type'] == 'ASSAULT':
        return 'BATTERY'
      elif a['Primary Type'] == 'KIDNAPPING':
        return 'BATTERY'
      elif a['Primary Type'] == 'ARSON':
        return 'BATTERY'
      elif a['Primary Type'] == 'INTIMIDATION':
        return 'BATTERY'
      elif a['Primary Type'] == 'DECEPTIVE PRACTICE':
        return 'DECEPTIVE PRACTICE'
      elif a['Primary Type'] == 'CONCEALED CARRY LICENCE VIOLATION':
        return 'WEAPON VIOLATION'
      elif a['Primary Type'] == 'CRIMINAL SEXUAL ASSAULT':
        return 'SEXUAL'
      elif a['Primary Type'] == 'SEX OFFENSE':
        return 'SEXUAL'
      elif a['Primary Type'] == 'HUMAN TRAFFICKING':
        return 'SEXUAL'
      elif a['Primary Type'] == 'PROSTITUTION':
        return 'SEXUAL'
      elif a['Primary Type'] == 'OFFENSE INVOLVING CHILDREN':
        return 'SEXUAL'
      elif a['Primary Type'] == 'NARCOTICS':
        return 'NARCOTICS'
      elif a['Primary Type'] == 'GAMBLING':
        return 'NARCOTICS'
      elif a['Primary Type'] == 'LIQUOR LAW VIOLATION':
        return 'NARCOTICS'
      elif a['Primary Type'] == 'OTHER NARCOTIC VIOLATION':
        return 'NARCOTICS'    
      elif a['Primary Type'] == 'PUBLIC PEACE VIOLATION':
        return 'NON-CRIMINAL'
      elif a['Primary Type'] == 'OBSCENITY':
        return 'NON-CRIMINAL'
      elif a['Primary Type'] == 'RITUALISM':
        return 'NON-CRIMINAL'
      elif a['Primary Type'] == 'PUBLIC INDECENCY':
        return 'NON-CRIMINAL'
      elif a['Primary Type'] == 'INTERFERENCE WITH PUBLIC OFFICER':
        return 'NON-CRIMINAL'
      elif a['Primary Type'] == 'NON-CRIMINAL':
        return 'NON-CRIMINAL'

crimes['Crime_Group'] = crimes.apply(crimegr, axis=1)
```

```python
crimes = crimes.drop(crimes[crimes['Community Area'] == 0].index)
crimes.head()
```

```python
crimes.pivot_table(crimes,index=["Community Area", "Primary Type"]))
crimes
```

# Reported Crimes - Community Area by type per year

```python
crime_data = crimes.pivot_table(values=["ID"], index=["Community Area", "Primary Type", "Year"], aggfunc=np.count_nonzero)
crime_data
```

```python
crime_data.to_csv("corrected_full_organized_crime_data.csv")
```

# Number of Reported Crimes per Commuity Area

```python
#1) group by community area
#2) get numbers of crime type per community area

crime_area= pd.DataFrame(crimes.groupby("Community Area")["ID"].count())
crime_area.head()
```

```python
crime_area.to_csv("corrected_full_crime_by_area.csv")
```

# Number of Reported Crimes per Primary Type from 2001-2018

```python
crime_type= pd.DataFrame(crimes.groupby("Primary Type")["ID"].count())
crime_type.sort_values(by= "ID", ascending= False)
```

```python
crime_type.to_csv("corrected_full_crime_by_type.csv")
```
